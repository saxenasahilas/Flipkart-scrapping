```python
import numpy as np
import pandas as pd
```

```python
a=pd.read_csv('flipkart final.csv')
a
```

```python
import matplotlib.pyplot as plt
```

```python
x=a['Price (in INR)']
y=a['Discount']
x=x.tolist()
x=[i.replace(',','') for i in x] 
x=[int(i) for i in x]
x.sort()
x
y=y.tolist()
y=[i.replace('% off','') for i in y] 
y=[int(i) for i in y]
y.sort()
y
plt.plot(x,y,c='r',)
plt.title('flpkrt',fontsize=20)
plt.show()
```

```python
d=a['Original Price Claimed(in INR)']
d=[int(i) for i in d]
d.sort()

plt.plot(y,d)
plt.plot(x,y,c='r',)
plt.plot(x,d)
plt.title('flpkrt',fontsize=20)
```
